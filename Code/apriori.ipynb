```python
import warnings
warnings.filterwarnings("ignore")  # oculta todos os warnings (inclusive DeprecationWarning)

import pandas as pd

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# 1. Base de transações
transacoes = [
    ['Leite', 'Pão', 'Manteiga'],
    ['Leite', 'Fralda', 'Cerveja', 'Ovos'],
    ['Pão', 'Manteiga', 'Fruta'],
    ['Leite', 'Pão', 'Fralda', 'Manteiga'],
    ['Pão', 'Fralda', 'Ovos'],
    ['Leite', 'Pão', 'Manteiga', 'Ovos'],
    ['Cerveja', 'Fralda', 'Pão'],
    ['Leite', 'Pão', 'Fralda', 'Cerveja'],
    ['Leite', 'Pão', 'Manteiga'],
    ['Pão', 'Fruta', 'Manteiga']
]
print("BASE DE TRANSAÇÕES:")

for i, t in enumerate(transacoes, start=1):
    print(f"T{i}: {t}")

# 2. Transformar em matriz One-Hot
te = TransactionEncoder()
df_bool = te.fit(transacoes).transform(transacoes)
df = pd.DataFrame(df_bool, columns=te.columns_)


print("\nMatriz One-Hot (itens x transações):")
display(df.head())

# 3. Rodar Apriori com suporte mínimo de 30%
min_support = 0.3
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

# Adicionar tamanho dos conjuntos (para separar L1, L2, L3)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)
print(f"\nItens frequentes encontrados (suporte mínimo = {min_support*100:.0f}%):")
display(frequent_itemsets.sort_values(by=['length', 'support'], ascending=[True, False]))

# 4. Mostrar conjuntos frequentes por tamanho
print("L1 (1-itemsets frequentes):")
display(frequent_itemsets[frequent_itemsets['length'] == 1])
print("L2 (2-itemsets frequentes):")
display(frequent_itemsets[frequent_itemsets['length'] == 2])

print("L3 (3-itemsets frequentes):")
display(frequent_itemsets[frequent_itemsets['length'] == 3])


# 5. Gerar regras de associação
regras = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
regras = regras.sort_values(by="confidence", ascending=False)

print("\nRegras de associação (confiança = 70%):")
display(regras[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# 6. Mostrar regras mais fortes
print("\nRegras mais fortes (ordenadas por confiança):")

for _, row in regras.iterrows():

    ant = ', '.join(list(row['antecedents']))
    cons = ', '.join(list(row['consequents']))
    print(f"| {ant} | {cons} | suporte={row['support']:.2f}, confiança={row['confidence']:.2f}, lift={row['lift']:.2f}")

```
